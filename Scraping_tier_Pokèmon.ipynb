```python
# Librerie
from bs4 import BeautifulSoup
from urllib.request import Request,urlopen
import pandas as pd
import string
import csv

# Importa il Dataset
data = pd.read_csv('pokemon_alopez247.csv')

# Settaggio parametri. Modifica le i due Nidoran per evitare errori con caratteri non-ascii
BASE_URL = "http://pokemon.neoseeker.com"
data.at[28,'Name'] = "Nidoran_(M)"
data.at[31,'Name'] = "Nidoran_(F)"

# Make BeautifulSoup4
def make_soup(url):
    req = Request(url, headers={'User-Agent': 'Dont Mind Me Im Just Scraping'})
    html = urlopen(req).read()
    return BeautifulSoup(html, "lxml")
```

```python
# Input: nome del Pokémon
# Output: Tier del Pokémon
def get_pokemon_tier(pokemon_name):
    pokemon_url = BASE_URL + '/wiki/' + pokemon_name
    soup = make_soup(pokemon_url)
    pokemon_data = []
    table_body = soup.find('table')

    rows = table_body.find_all('tr')
    for row in rows:
        col = row.find('td').text.strip()
        pokemon_data.append(col)
    return pokemon_data[16]
```

```python
# Crea una lista con i tier dei primi 667 Pokémon
pokelist = []
for index,row in data[:668].iterrows():
    pokelist.append(get_pokemon_tier(row['Name']))
    print(row['Name'], pokelist[-1], len(pokelist))
```

```python
# Il Pokémon num. 668 (Flabébé) deve essere inserito manualmente.
pokelist.append('NFE')
# Scraping degli ultimi Pokémon
for index,row in data[669:].iterrows():
    pokelist.append(get_pokemon_tier(row['Name']))
    print(row['Name'], pokelist[-1], len(pokelist))
```

```python
# Creazione dictionary con Numero, Tier
pokedict = dict(zip(range(1,722),pokelist))
pokedict
```

```python
# Scrittura su disco del file csv
download_dir = "tier_list.csv"

csv = open(download_dir, "w") 

columnTitleRow = "Number, Tier\n"
csv.write(columnTitleRow)

for key in pokedict.keys():
    number = key
    tier = pokedict[key]
    row = str(number) + "," + tier + "\n"
    csv.write(row)
```

```python
# Pulizia dei dati
import re
pokedict_clean = pokedict

for key in pokedict_clean:
    temp = re.findall('[a-zA-Z0-9]+\Z',pokedict_clean[key])[0]
    if temp == 'BL':
        temp = 'BL1'
    if temp == 'LC':
        temp = 'NFE'
    if temp == 'Unknown':
        temp = 'TBD'
    pokedict_clean[key] = temp
```

```python
# Modifica i cluster in numeri

pokechange = {'Uber' : '1', 'OU' : '2', 'BL1' : '3', 'UU' : '4', 'BL2' : '5', 'RU' : '6', 'BL3' : '7', 'NU' : '8', 'BL4' : '9', 'PU' : '10', 'NFE' : '11', 'TBD' : '12'}

def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text

for key in pokedict_clean:
    pokedict_clean[key] = replace_all(pokedict_clean[key], pokechange)
```

```python
print(set(list(pokedict_clean.values())))
print('Numero di cluster:',len(set(list(pokedict_clean.values()))))
```

```python
# Scrittura su disco del file csv aggiornato
download_dir = "tier_list_fixed.csv"

csv = open(download_dir, "w") 

columnTitleRow = "Number,Tier\n"
csv.write(columnTitleRow)

for key in pokedict_clean.keys():
    number = key
    tier = pokedict_clean[key]
    row = str(number) + "," + tier + "\n"
    csv.write(row)
```
